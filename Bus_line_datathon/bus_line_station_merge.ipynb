```python
import pandas as pd
pd.options.display.max_rows = 1000
```

```python
df_route_station = pd.read_csv("./data/PJT001_routestationinfo.csv")
df_route_station["station_id"] = df_route_station["station_id"].fillna(0.0).astype(int)
df_route_station["mobile_no"] = df_route_station["mobile_no"].fillna(0.0).astype(int)
df_route_station.head(30)
```

```python
df_station = pd.read_csv("./data/PJT001_stations_table.csv")
df_station.head(500)
```

```python
df_station_for_join = df_station.loc[:, ["정류소명", "정류소번호", "모바일정류장ID", "WGS84위도", "WGS84경도"]]
df_station_for_join.columns = ["station_nm", "station_id", "mobile_no", "lon", "lat"]

for i, val in enumerate(df_station_for_join["mobile_no"]):
    try:
        df_station_for_join.loc[i, "mobile_no"] = float(val)
    except:
        df_station_for_join.loc[i, "mobile_no"] = 0


df_station_for_join["station_id"] = df_station_for_join["station_id"].fillna(0.0).astype(int)
df_station_for_join["mobile_no"] = df_station_for_join["mobile_no"].fillna(0.0).astype(int)
df_station_for_join.head()
```

```python
df_station_drop_dup = df_station_for_join.drop_duplicates(["station_nm"], keep="first")
```

```python
test = pd.merge(df_route_station, df_station_drop_dup, how = "inner", on = "station_id")
test.head()
```

```python
test = pd.merge(df_route_station, df_station_drop_dup, how = "inner", on = "station_nm")
test.head(100)
```

```python
test[test["bus_line_no"] == "10-4"].sort_values(["pr_station_id", "bus_line_no_seq"])
```

```python
bus_line_station_merge = pd.merge(df_route_station, df_station_drop_dup, how = "inner", on = "station_nm")
bus_line_station_merge = bus_line_station_merge.sort_values(["pr_station_id", "bus_line_no_seq"])
bus_line_station_merge.head(500)
```
