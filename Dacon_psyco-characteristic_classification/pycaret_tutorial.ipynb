```python
import os
import pandas as pd
from pycaret.classification import *

path = './data/'

train      = pd.read_csv(path + "train.csv").iloc[:, 1:]
test       = pd.read_csv(path + "test_X.csv").iloc[:, 1:]
submission = pd.read_csv(path + "sample_submission.csv")
```

```python
train.shape
```

```python
test.shape
```

```python
clf = setup(data=train, target='voted')
```

```python
best_5 = compare_models(sort='AUC', n_select=5)
```

```python
ensemble_model = blend_models(estimator_list=best_5, 
                             fold=5,
                             method='soft')
```

```python
pred = predict_model(ensemble_model)
```

```python
final_model = finalize_model(ensemble_model)
```

```python
pred = predict_model(final_model, data=test)
pred
```

```python
submission['voted'] = pred['Score']
submission.to_csv('submission_proba.csv', index=False)
```
